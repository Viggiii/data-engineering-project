```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import csv
import pandas as pd
```

```python
def check_price():
  # to connect to the amazon website 
  URL= 'https://www.amazon.com/s?k=laptops&crid=O59B07JDUOK&sprefix=laptops%2Caps%2C711&ref=nb_sb_ss_ts-doa-p_1_7'
  #to sent HTTP request  and receive response from server
  headers= {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}

  # to get the page from the website

  page= requests.get(URL,headers=headers)

  soup1=BeautifulSoup(page.content,  'html.parser')
  # prettify gives the content readable 
  soup2=BeautifulSoup(soup1.prettify(),'html.parser')
  # Extract the title of the product from the website
  title = soup2.find('span', attrs={'class':'a-size-medium a-color-base a-text-normal'}).get_text().strip()
  # Extract the product price from the website
  price = soup2.find('span', attrs={'class':'a-price-whole'}).get_text().strip()
  # Extract the Rating from the each product
  rating =soup2.find('span',attrs={'class':'a-icon-alt'}).get_text().strip()
  #to get the date  and time when the script was run
  today=datetime.date.today()

  #to give header to the products
  header=['title','price','rating','Date']
  #conversion string to list
  data = [title,price,rating,today]
  
  # appending the data into the csv

  with open('AmazonwebscrapingDataset.csv','a+',newline='',encoding='UTF8') as f:
    writer=csv.writer(f)
    writer.writerow(data)
```

```python

#to set the timer and extract the data  from the website
while(True):
  check_price()
  time.sleep(2)

  
```
